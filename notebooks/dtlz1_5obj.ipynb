# DTLZ1 Results

```python
import sys
import os

# Adiciona a pasta raiz do projeto aos caminhos que o Python reconhece
sys.path.append(os.path.abspath('..'))

from functools import partial
from src.meamt_core import build_toolbox, gen_inicial_tables, run, generate_dtlz1_front_random
from deap import benchmarks
import numpy as np
os.makedirs("Resultados_dtlz1_5obj", exist_ok=True)

# 1. Parâmetros
K = 5
NPOP = 1600
NGEN = 600
NOBJ = 5 
IND_SIZE = NOBJ + K - 1 
NUM_TABLES = int((1 << NOBJ)) 
MAX_TABLE_SIZE = int(NPOP / NUM_TABLES)
REF_POINT_HV = [1.1]*NOBJ
RESET = 150
CXPB = 0.9
MUTPB = 1.0

func_dtlz1_segura = partial(benchmarks.dtlz1, obj=NOBJ)
toolbox = build_toolbox(func_dtlz1_segura, IND_SIZE, NPOP, NOBJ)

for i in range(1, 31):
    if __name__ == '__main__':
        pop_inicial = toolbox.population()
        
        fitnesses = toolbox.map(toolbox.evaluate, pop_inicial)
        
        for ind, fit in zip(pop_inicial, fitnesses):
            ind.fitness.values = fit
            
        tabelas = gen_inicial_tables(pop_inicial, NUM_TABLES, MAX_TABLE_SIZE, NOBJ)
        pareto_real = generate_dtlz1_front_random(NOBJ, 1000)
        
        logbook = run(tabelas, pareto_real, NUM_TABLES, MAX_TABLE_SIZE, NGEN, toolbox, CXPB, MUTPB, REF_POINT_HV, NOBJ, RESET)
        
        fronteira_final = [ind.fitness.values for ind in tabelas[0]]
        matriz_resultados = np.array(fronteira_final) 
        nome_arquivo = f"Resultados_dtlz1_5obj/meamt_run_{i}.npy"
        np.save(nome_arquivo, matriz_resultados)  
```
